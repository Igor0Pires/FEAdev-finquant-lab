#### **Dicas**



**Dica geral:** Sempre que possível, use as propriedades dos momentos para simplificar cálculos. Lembre-se:
- Valor esperado é linear: E[aX + bY] = aE[X] + bE[Y]
- Variância não é linear: Var(aX) = a²Var(X)
- Skewness e curtose mudam ou se mantêm dependendo da transformação

**Para verificação:** 
- Índice de Sharpe = (Retorno - Taxa livre de risco) / Volatilidade
- Covariância(X,Y) = Correlação(X,Y) × σ_X × σ_Y
- Variância de portfólio: Var(wX + (1-w)Y) = w²Var(X) + (1-w)²Var(Y) + 2w(1-w)Cov(X,Y)

**Interpretações importantes:**
- Skewness negativa → mais perdas extremas que ganhos extremos
- Curtose > 3 → mais eventos extremos que distribuição normal
- Correlação < 1 → benefícios da diversificação
